```python
# Input sentence
sent = input("Enter sentence: ")
# splitting the sentence into words
sent = sent.split()
# declaring the empty dictionary
words = {}
# looping through the each word in the sentence
for word in sent:
    # Incrementing the word occurence into the dictionary
    words[word] = words.get(word, 0) + 1

# Printing the how many unique words
print(f"Unique words count: {len(words)}")
# printing the unique word
print(f"Unique words: {tuple(words.keys())}")
```

```python
# Input employees
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# calculating the max salary among the employees
highest_salary = max(employees.values())

# looping through the employees
for employee, salary in employees.items():
    # checking the salary is equal to the highest salary amon employees
    if salary==highest_salary:
        # if saalry is equal to highest salary print the employee and salary
        print(f"Highest Salary: {employee} - {highest_salary}")
```

# 3.

```python
# Input list
li = eval(input("numbers = "))
# len of the li
l = len(li)
# assigning the first element of the list to max_ and min_
max_ = li[0]
min_ = li[0]
# looping through the numbers in list
for num in li[1:]:
    # changing the max_ value if the current num is greather than max_
    if num > max_:
        max_ = num
    # changing the min_ value if the current element is less than min_
    if num < min_:
        min_ = num
# printing the list of numbers
print(f"List: {li}")
# printing the max_ number
print(f"Max: {max_}")
# printing the min_ number
print(f"Min: {min_}")
```

```python
# input list
li = eval(input("prices = "))
# assiging the count variable to 0 for counting
count = 0
# looping through the list of prices
for price in li:
    if price>1000:
        # If price is greater than 1000, increment the count variable
        count+=1
# printing the no.of products which has greater than 1000
print(f"Products above 1000: {count}")
```
