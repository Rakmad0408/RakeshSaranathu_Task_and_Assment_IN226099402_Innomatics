```python
def recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount in valid_plans:
        return True
    return False
```

```python
while True:
    recharge_amount = int(input("Enter the recharge amount: "))
    if recharge_amount>=50:
        if recharge(recharge_amount):
            print("Recharge Successful")
            break
        else:
            print("Kindly enter valid amount and retry!!!")
    else:
        print("Minimum amount should be ₹50")
```

```python
def alert_system(stock):
    if stock<15:
        print("Reorder Alert")
    else:
        print("Stock Ok")
```

```python
dict_ = {
    "Shirt": 15,
    "Pant": 3
}

for i in dict_:
    alert_system(dict_[i])
```

# 3.

```python
def result_processing_system(marks):
    avg_marks = sum(marks)/len(marks)
    print("Pass") if avg_marks>=50 else print("Fail")
```

```python
marks = [[10, 20, 30, 40], [50, 60, 70, 80]]
for i in marks:
    result_processing_system(i)
```

```python
import time
```

```python
def cab_fair_estimator(km, base=50):
    total_fare = km*12+base
    if 4<=time.localtime()[3]%12<=9:
        total_fare = total_fare*1.25
    return total_fare
```

```python
count = 0
while True:
    km = int(input("Enter total km: "))
    print(cab_fair_estimator(km))
    count+=1
    if count == 10:
        break
```

```python
def attendance_eligibility(li):
    p_count = li.count("P")
    per = p_count/len(li)
    if per >= 75:
        print("Eligible")
    else:
        print("Not Eligible")
```

```python
def strength_checker(password):
    len_8 = True if len(password)>=8 else False
    is_digit = any(i.isdigit() for i in password)
    is_char = any(True for i in password if i in string.punctuation)
    if all(len_8, is_char, is_digit):
        print("strong Pa")
```
