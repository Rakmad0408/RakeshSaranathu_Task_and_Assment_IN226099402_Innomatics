```python
def analyze_rainfall(rainfall_data, required_level):
    total = 0
    # Calculate total rainfall using loop
    for rain in rainfall_data:
        total += rain
    # Calculate average
    average = total / len(rainfall_data)
    # Check rainfall status
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    # Print output
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)
```

```python
rainfall = [70, 75, 80, 65, 70]
required = 70

analyze_rainfall(rainfall, required)
```

```python
def detect_duplicate_usernames(usernames):
    unique_users = set()
    duplicate_found = False
    
    for user in usernames:
        if user in unique_users:
            duplicate_found = True
            break
        unique_users.add(user)
    
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
```

```python
user_list = ["rakesh", "john", "sara", "rakesh", "mike"]

detect_duplicate_usernames(user_list)
```

```python
def filter_premium_crops(prices):
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)
```

```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    # Count present days using loop
    for day in attendance_list:
        if day == 1:   # 1 means Present, 0 means Absent
            present_days += 1
    # Calculate attendance percentage
    percentage = (present_days / total_days) * 100
    # Check eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)
```

```python
attendance = [1, 1, 1, 0, 1]   # 4 out of 5 days present â†’ 80%
check_exam_eligibility(attendance)
```

```python
def calculate_electricity_bill(units):
    bill = 0
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status
```

```python
units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
